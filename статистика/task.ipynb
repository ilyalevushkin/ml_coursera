```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
from math import sqrt
```

```python
expon_rv = sts.expon()
sample = expon_rv.rvs(size=1000)
```

```python
plt.hist(sample, density=True)

x = np.linspace(0,9,1000)
pdf = expon_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

```python
def sample_generate(n, count, rv):
    L = []
    for i in range(count):
        L.append(rv.rvs(size=n))
    return L

d = {}

d.update({5 : sample_generate(5, 1000, expon_rv)})
d.update({10: sample_generate(10, 1000, expon_rv)})
d.update({50: sample_generate(50, 1000, expon_rv)})
```

```python
def calc_average(lst):
    summ = 0
    for i in lst:
        summ += i
    return summ / float(len(lst))

d_sample = {}

d_sample.update({5 : [calc_average(i) for i in d[5]]})
d_sample.update({10 : [calc_average(i) for i in d[10]]})
d_sample.update({50 : [calc_average(i) for i in d[50]]})
```

# Выкладки с вычислениями параметров нормальных распределений, аппроксимирующих выборочные средние при различных $n$;

Мат. ожидание экспоненциальной случайной величины равна $\frac{1}{\lambda}$, а ее дисперсия $\frac{1}{\lambda^2}$.

Поскольку мы приняли $\lambda = 1$:

$MX = 1$

$DX = 1$

Соответственно, параметры нормальных распределений для выборочного среднего этой величины:

$\mu = 1$

$\sigma^2 = \frac{1}{n}$

```python
plt.hist(d_sample[5], density=True, label='n = 5')


x = np.linspace(0,4,1000)
norm_rv = sts.norm(1, sqrt(1/float(5)))
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

```python
plt.hist(d_sample[10], density=True, label='n = 10')


norm_rv = sts.norm(1, sqrt(1/float(10)))
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

```python
plt.hist(d_sample[50], density=True, label='n = 50')


norm_rv = sts.norm(1, sqrt(1/float(50)))
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

# Выводы

С ростом $n$ точность апроксимации выборочных средних нормальным увеличивается. Стоит заметить, что при увеличении объема выборки, дисперсия выборочного среднего уменьшается (стремится к 0), следовательно все большая часть площади под графиком плотности выборочного среднего становится заключена в меньшем интервале от 1.
